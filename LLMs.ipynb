```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import torch
import sacrebleu

# Load mBART-50 model and tokenizer
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Function to translate text using mBART with forced target language token
def translate_with_mbart(text, source_lang="en_XX", target_lang="fr_XX"):
    tokenizer.src_lang = source_lang

    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt").to(device)

    # Generate the translation, explicitly specifying the target language
    forced_bos_token_id = tokenizer.lang_code_to_id[target_lang]
    translated_tokens = model.generate(
        inputs["input_ids"],
        max_length=512,
        num_beams=4,
        early_stopping=True,
        forced_bos_token_id=forced_bos_token_id  # Force the model to generate in the target language
    )

    # Decode the translated tokens
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
    return translated_text

# Function to read input text from a file and translate it
def translate_file(input_file, output_file, source_lang="en_XX", target_lang="fr_XX"):
    with open(input_file, 'r', encoding='utf-8') as file:
        input_text = file.read()  # Read the entire file content

    # Translate the text
    translated_text = translate_with_mbart(input_text, source_lang=source_lang, target_lang=target_lang)

    # Write the translated text to the output file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(translated_text)

    print(f"Translated text written to {output_file}")
    return translated_text

# Function to evaluate translation quality using BLEU score
def evaluate_translation(translations, references):
    bleu = sacrebleu.corpus_bleu(translations, [references])
    print(f"BLEU Score: {bleu.score}")
    return bleu.score

# Example usage
input_file = "/content/sample_data/Script.txt"  # Path to the input file
output_file = "/content/sample_data/Output-french.txt"  # Path to the output file

# Translate from English to French
translated_text = translate_file(input_file, output_file, source_lang="en_XX", target_lang="fr_XX")

# Reference (ground truth) translations for evaluation (replace with your actual references)
reference_translations = [
    "Le film RRR est une grande représentation du cinéma indien.",
    "Ceci est un exemple d'évaluation de la qualité de la traduction.",
    "Le renard brun rapide saute par-dessus le chien paresseux."
]

# Evaluate the translated text against the references
evaluate_translation([translated_text], reference_translations)
```

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast, AutoModelForSeq2SeqLM, AutoTokenizer, pipeline
import sacrebleu

# Load mBART and NLLB models
def load_mbart_model():
    model_name = "facebook/mbart-large-50-many-to-many-mmt"
    tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
    model = MBartForConditionalGeneration.from_pretrained(model_name)
    return model, tokenizer

def load_nllb_model():
    model_name = "facebook/nllb-200-distilled-600M"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    return model, tokenizer

# Translation functions for each model
def translate_with_mbart(text, source_lang="en_XX", target_lang="hi_IN"):
    model, tokenizer = load_mbart_model()
    tokenizer.src_lang = source_lang
    inputs = tokenizer(text, return_tensors="pt")
    forced_bos_token_id = tokenizer.lang_code_to_id[target_lang]
    translated_tokens = model.generate(
        inputs["input_ids"],
        max_length=512,
        num_beams=4,
        early_stopping=True,
        forced_bos_token_id=forced_bos_token_id
    )
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
    return translated_text

def translate_with_nllb(text, source_lang="eng_Latn", target_lang="hin_Deva"):
    model, tokenizer = load_nllb_model()
    translator = pipeline('translation', model=model, tokenizer=tokenizer, src_lang=source_lang, tgt_lang=target_lang, max_length=512)
    translated_text = translator(text)[0]['translation_text']
    return translated_text

# Function to evaluate the translation using BLEU score
def evaluate_bleu(translated_text, reference_text):
    bleu = sacrebleu.corpus_bleu([translated_text], [[reference_text]])
    return bleu.score

# Function to compute Mean Win Rate
def compute_mean_win_rate(scores):
    num_models = len(scores)
    win_matrix = [[0] * num_models for _ in range(num_models)]

    # Fill in the win matrix by comparing BLEU scores
    for i in range(num_models):
        for j in range(num_models):
            if i != j:
                win_matrix[i][j] = 1 if scores[i] > scores[j] else 0

    # Calculate the win rate for each model
    win_rates = [sum(win_matrix[i]) / (num_models - 1) for i in range(num_models)]
    return win_rates

# Example usage
def main():
    # Sample text and reference translations (replace with actual data)
    input_text = "RRR movie is a great example of Indian cinema."
    reference_translation = "आरआरआर फिल्म भारतीय सिनेमा का एक उत्कृष्ट उदाहरण है।"

    # Translate using mBART
    mbart_translation = translate_with_mbart(input_text)

    # Translate using NLLB
    nllb_translation = translate_with_nllb(input_text)

    # Evaluate BLEU scores for both models
    mbart_bleu = evaluate_bleu(mbart_translation, reference_translation)
    nllb_bleu = evaluate_bleu(nllb_translation, reference_translation)

    # Store BLEU scores
    bleu_scores = [mbart_bleu, nllb_bleu]

    # Compute Mean Win Rate (MWR)
    win_rates = compute_mean_win_rate(bleu_scores)

    # Print results
    model_names = ["mBART", "NLLB"]
    for i, win_rate in enumerate(win_rates):
        print(f"Model: {model_names[i]}, BLEU Score: {bleu_scores[i]:.2f}, Mean Win Rate: {win_rate:.2f}")

if __name__ == "__main__":
    main()
```
